```python
def NN(config,i,j,L):
#nearest  neighbours   
  
        return [[(i+1)%L,j],[i,(j-1)%L],[(i-1)%L,j],[i,(j+1)%L]]
def nearest(grid,i,j,L):
#spin of nearest neighbours    
    q1=[]
    q=NN(grid,i%L,j%L,L)
    for i in q:
        if i!=0 :
           
            q1.append(grid[i[0],i[1]])
        else:
            q1.append(0)
    return q1            
```

```python
a= np.ones([3,3])
a[1,2]=2
#co-ordiante of interfacial cells in general ()
def rinterfacial(grid,i,j,L): #i j, is center of template and L size of grid 
    q1=interfacial(grid,i,j,L)
    
    q1=[y for y in q1 if y[1]>j]
    
    
    return q1
def interfacial(grid,i,j,L):
    q=NN(grid,i,j,L)
    q1=[]
    for k in q:
        q1+=NN(grid,k[0],k[1],L)
    
    q1= duplicate(q1)
    if [i,j] in q1:
        q1.remove([i,j])
    return q1
def linterfacial(grid,i,j,L):
    a=interfacial(grid,i,j,L)
    c=[y for y in a if y[1]<j]
    return c
def topinterfacial(grid,i,j,L):
    
    return [[i-2,j]]
def botinterfacial(grid,i,j,L):
  
    return [[i+2,j]]

    
def duplicate(a): # list duplicating
    b=[]
    for i in a:
        if i not in b:
            b.append(i)
    return b
```
